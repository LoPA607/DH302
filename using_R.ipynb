```r
# Install the tidyverse package if you haven't already
install.packages("tidyverse")
```

```r
df_grip<-read_tsv("/content/Assignment02_grip_strength.tsv")
```

```r
df_grip <- df_grip %>%
  mutate(diff = measured_6weekslater - baseline)

ggplot(df_grip, aes(x = diff)) +
  geom_histogram(binwidth = 1, fill = "lightblue", color = "black") +
  labs(title = "(After - Before)",
       x = "Difference", y = "Count") +
  theme_minimal()

qqnorm(df_grip$diff)
qqline(df_grip$diff, col = "red")
```

```r
t_test_result <- t.test(df_grip$baseline, df_grip$measured_6weekslater,
                        paired = TRUE, alternative = "less")
t_test_result$p.value
```

```r
wilcox_result <- wilcox.test(df_grip$baseline, df_grip$measured_6weekslater,
                             paired = TRUE, alternative = "less")
wilcox_result$p.value
```
